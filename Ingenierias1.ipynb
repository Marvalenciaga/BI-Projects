```python
import openpyxl
import pandas as pd

#Leer excel
df= pd.read_excel("POS_MED_MAN.xlsx", sheet_name="Hoja1")
print(df.head)

#Analisis de variables
Datos=pd.pivot_table(df, values="PAPA", index=["NIVEL", "FACULTAD"], columns="SEDE", aggfunc="mean", fill_value=0)
print(Datos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

Datos.describe()
Datos.plot(kind='barh')
plt.show()
```

```python
Medellin=Datos[["SEDE MEDELLÍN"]]
print(Medellin)
```

```python
Medellin.describe()
Medellin.plot(kind='barh')
plt.show()
```

```python
Datos2=pd.pivot_table(df, values="FACULTAD", index=["NIVEL"], columns="SEDE", aggfunc="count", fill_value=0)
print(Datos2)
```

```python
Datos2.describe()
Datos2.plot(kind='barh')
plt.show()
```

```python
Datos3=pd.pivot_table(df, values="PLAN", index=["FACULTAD", "NIVEL"], columns="SEDE", aggfunc="count", fill_value=0)
print(Datos3)
```

```python
Datos3.describe()
Datos3.plot(kind='barh')
plt.show()
```
