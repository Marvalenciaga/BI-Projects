```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols

# Cargar los datos desde un archivo Excel
file_path = "/content/Engineering_anonimizado.xlsx"  # Reemplaza con la ruta de tu archivo
data = pd.read_excel(file_path)

# 1. Inspección inicial
print("Primeras filas del dataset:")
print(data.head())
print("\nInformación del dataset:")
print(data.info())

# 2. Convertir la columna SEMESTER en formato ordenado para el eje X
def ordenar_semestre(semestre):
    # Split the semester string using regular expression to handle different formats
    import re
    match = re.match(r'(\d+)-(\d+)([a-zA-Z]?)', semestre)
    if match:
        año, periodo, suffix = match.groups()
        # Convert year and period to integers, handling optional suffix
        return int(año) * 2 + int(periodo) #+ (0 if suffix is None or suffix == '' else 1)
    else:
        # Handle cases where the semester string doesn't match the expected format
        return -1  # Or any other appropriate value to indicate an error

# Asegúrate de que la columna 'SEMESTER' esté en formato string
data['SEMESTER'] = data['SEMESTER'].astype(str)

# Crear una nueva columna para ordenar los semestres
data['SEMESTER_ORDERED'] = data['SEMESTER'].apply(ordenar_semestre)

# Ordenar los datos por semestre
data = data.sort_values(by='SEMESTER_ORDERED')

# 3. Gráfico de evolución del rendimiento académico por semestre
sns.lineplot(data=data, x='SEMESTER', y='CWAA', hue='SOCIOECONOMIC_ESTATUS')
plt.xticks(rotation=90)  # Rotar etiquetas del eje X para mejor legibilidad
plt.title("Rendimiento Académico por Semestre")
plt.xlabel("Semestre")
plt.ylabel("Promedio Académico (CWAA)")
plt.legend(title="Nivel Socioeconómico")
plt.show()

sns.lineplot(data=data, x='SEMESTER', y='CWAA', hue='GENDER', palette=["#2a9d8f", "#e76f51"])
plt.xticks(rotation=90)
plt.title("Rendimiento Académico por Semestre y Género")
plt.xlabel("Semestre")
plt.ylabel("Promedio Académico (CWAA)")
plt.legend(title="Género", labels=["Masculino", "Femenino"])
plt.show()

# 5. Gráfico de evolución por programa académico
sns.lineplot(data=data, x='SEMESTER', y='CWAA', hue='PROGRAM')
plt.xticks(rotation=90)
plt.title("Rendimiento Académico por Semestre y Programa")
plt.xlabel("Semestre")
plt.ylabel("Promedio Académico (CWAA)")
plt.legend(title="Programa Académico")
plt.show()
```
