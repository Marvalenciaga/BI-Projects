```python
class SocioeconomicLevel:
    def __init__(self, SOCIOECONOMIC_ESTATUS):
        self.SOCIOECONOMIC_ESTATUS = SOCIOECONOMIC_ESTATUS

class Gender:
    def __init__(self, GENDER):
        self.GENDER = GENDER

class Program:
    def __init__(self, PROGRAM):
        self.PROGRAM = PROGRAM

class AverageScore:
    def __init__(self, CWAA):
        self.CWAA = CWAA

class Age:
    def __init__(self, AGE):
        self.AGE = AGE

class Faculty:
    def __init__(self, STUDENT, GENDER, PROGRAM, SOCIOECONOMIC):
        self.STUDENT = STUDENT
        self.AverageScore = AverageScore
        self.Gender = Gender
        self.Program = Program
        self.SocioeconomicLevel = SocioeconomicLevel
        self.Age = Age
        self.Faculty = Faculty

class Student:
    def __init__(self, SOCIOECONOMIC, PROGRAM, AVERAGE_SCORE, GENDER, AGE):
        self.SOCIOECONOMIC = SOCIOECONOMIC
        self.PROGRAM = PROGRAM
        self.AVERAGE_SCORE = AVERAGE_SCORE
        self.GENDER = GENDER
        self.AGE = AGE
        self.Faculty = Faculty

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('/content/Engineering_anonimizado.xlsx')
print(df.head())

# 1. Average for CWAA and Headquarters
average_for_headquarters = df.groupby('CAMPUS')['CWAA'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='CAMPUS', y='CWAA', data=average_for_headquarters, palette='viridis')
plt.title('Average score for Campus')
plt.xlabel('Campus')
plt.ylabel('Average Score (CWAA)')
plt.xticks(rotation=45, ha="right")
plt.show()

# 2. Average for CWAA and Socioeconomic_Estatus
average_for_socioeconomic_estatus = df.groupby('SOCIOECONOMIC_ESTATUS')['CWAA'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='SOCIOECONOMIC_ESTATUS', y='CWAA', data=average_for_socioeconomic_estatus, palette='Blues_d')
plt.title('Average score for estatus')
plt.xlabel('Socioeconomic Estatus')
plt.ylabel('Average Score (CWAA)')
plt.xticks(rotation=45, ha="right")
plt.show()

# 3. Average for CWAA and Gender
average_for_gender = df.groupby('GENDER')['CWAA'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='GENDER', y='CWAA', data=average_for_gender, palette='Set2')
plt.title('Average score for gender')
plt.xlabel('Gender')
plt.ylabel('Average Score (CWAA)')
plt.show()

# 4. Average for CWAA and Program
average_for_program = df.groupby('PROGRAM')['CWAA'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='PROGRAM', y='CWAA', data=average_for_program, palette='coolwarm')
plt.title('Average score for Program')
plt.xlabel('Program')
plt.ylabel('Average Score (CWAA)')
plt.xticks(rotation=45, ha="right")
plt.show()

# 5. Average for CWAA and Faculty
if 'FACULTY' in df.columns:
    average_for_faculty = df.groupby('FACULTY')['CWAA'].mean().reset_index()

    plt.figure(figsize=(10, 6))
    sns.barplot(x='FACULTY', y='CWAA', data=average_for_faculty, palette='PuBuGn_d')
    plt.title('Average score for Faculty')
    plt.xlabel('Faculty')
    plt.ylabel('Average Score (CWAA)')
    plt.xticks(rotation=45, ha="right")
    plt.show()

# 6. Average for CWAA and Age range
bins = [18, 25, 30, 35, 40, 50, 60]
labels = ['18-25', '26-30', '31-35', '36-40', '41-50', '51-60']

df['Age_range'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=False)

average_for_age_range = df.groupby('Age_range')['CWAA'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Age_range', y='CWAA', data=average_for_age_range, palette='OrRd')
plt.title('Average score for age range')
plt.xlabel('Age Range')
plt.ylabel('Average Score (CWAA)')
plt.show()

```

```python
import pandas as pd
import hashlib

seed = "My_seed_unique"

def hash_value(value, seed):
    if pd.isnull(value):
        return None
    value_with_seed = f"{seed}{value}"
    return hashlib.sha256(value_with_seed.encode('utf-8')).hexdigest()

file_path = "/content/archivo_admitidos.xlsx"
df = pd.read_excel(file_path)

print("Original data:")
print(df.head())

columns_to_hash = ['ID']

for column in columns_to_hash:
    df[column] = df[column].apply(lambda x: hash_value(x, seed))

output_file_path = '/content/Data_Admitted.xlsx'
df.to_excel(output_file_path, index=False)
```
